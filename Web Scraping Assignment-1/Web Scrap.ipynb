## 1) Write a python program to display all the header tags from wikipedia.org and make data frame.


```python
from bs4 import BeautifulSoup
import pandas as pd
import warnings 
import requests
warnings.filterwarnings("ignore")
```

```python
url = "https://en.wikipedia.org/wiki/Main_Page"
```

```python
wiki_page = requests.get(url)
```

```python
wiki_page
```

```python
wiki_content = BeautifulSoup(wiki_page.content,"html.parser")
```

```python
wiki_content
```

```python
header = ['h1','h2','h3','h4','h5','h6']
```

```python
header_tag =[]
header_text= []

for tags in wiki_content.find_all(header):
    header_tag.append(tags.name)
    header_text.append(tags.text)
    
```

```python
print(len(header_tag),len(header_text))
```

```python
for i in range(len(header_tag)):
    x = header_tag[i]
    y = header_text[i]
    print(x,"->",y)
```

```python
df = pd.DataFrame({'Header Tag':header_tag,
                  'Header Text': header_text})
```

```python
df
```
