## Q1 Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.
## Q2 In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
```

```python
product = input("Enter a product: ")

driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
    
driver.get("http://www.amazon.in/")
time.sleep(2)
    
search_box = driver.find_element("id",'twotabsearchtextbox')
search_box.send_keys(product)

search_button = driver.find_element("id",'nav-search-submit-button')
search_button.click()
    
time.sleep(2)
opening_url = []
start = 0
end= 3
for page in range(start,end):
    time.sleep(3)
    try:
        url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
        for i in url:           
            opening_url.append(i.get_attribute('href'))
        next_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        next_button.click()
    except NoSuchElementException as e:
            print(f"Page not found{e}")

brand_name = []
name_of_the_product = []
price_product = []
expected_delivery = []
availability = []
product_url = []

for i in opening_url:
    
    driver.get(i)
    time.sleep(2)
    try:
        name = driver.find_element(By.XPATH,'//span[@class="a-size-base po-break-word"]')
        brand_name.append(name.text)
    except NoSuchElementException:
        brand_name.append('-')
        
    try:
        product_name = driver.find_element("id",'productTitle')
        name_of_the_product.append(product_name.text)
    except NoSuchElementException:
        name_of_the_product.append('-')
        
    try:
        price = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price_product.append(price.text)
    except NoSuchElementException:
        price_product.append('-')
        
    try:
        available = driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        availability.append(available.text)
    except NoSuchElementException:
        availability.append('-')
        
    try:
        product_url.append(i)
    except NoSuchElementException:
        product_url.append('-')
        
    try:
        expected_date = driver.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        expected_delivery.append(expected_date.text)
    except NoSuchElementException:
        expected_delivery.append('-')
    
driver.quit()
    
```

```python
print(len(brand_name),len(name_of_the_product),len(price_product),len(product_url),len(availability),len(expected_delivery))
```

```python
df = pd.DataFrame({"Brand Name":brand_name,
                  "Name of The Product": name_of_the_product,
                  "Price Of The Product": price_product,
                  "Availability":availability,
                  "Expected Delivery": expected_delivery,
                  "Product URL": product_url})
```

```python
df.to_csv('product_info.csv',index= False)
```

## Q3  Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 

```python
keywords = ['fruits','cars','Machine Learning', 'Guitar', 'Cakes']


for keyword in keywords:
    time.sleep(2)
    driver.get('https://images.google.com/')

    google_search = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
    google_search.send_keys(keyword)

    google_click = driver.find_element(By.XPATH,'//button[@class="Tg7LZd"]')
    google_click.click()

    img_tabs = []

    for page in range(0,1):
        url = driver.find_elements(By.XPATH,'//img[@class="rg_i Q4LuWd"]')
        count = 0
        for i in url[:10]:
            image_url = i.get_attribute('src')
            img_tabs.append(image_url)
            urllib.request.urlretrieve(image_url, f"{keyword}_{count}.jpg")
            count += 1

    time.sleep(1)
    
    
driver.quit()
   
```

```python
#“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 
driver = webdriver.Chrome(r"C:\Program Files (x86)\chromedriver.exe")
driver.get("https://www.forbes.com/")


time.sleep(10)


side_bar = driver.find_element(By.XPATH,'//div[@class="_69hVhdY4"]')

side_bar.click()

time.sleep(3)

billi = driver.find_element(By.XPATH,'//div[@class="mpBfVZz3"]')
action_chains = ActionChains(driver)
action_chains.move_to_element(billi).perform()


time.sleep(1)

world_billi = driver.find_element(By.XPATH,'//li[@class="TjJgrPSg _2bNo56RE secondary"]//a')
world_billi.click()

time.sleep(3)

name_of_person = []
rank_of_person = []
net_worth_person = []
age_person = []
citizenship_person = []
source_person = []
industry_person =[]


start = 0
end = 14
try:
    for page in range(start,end):
        time.sleep(1)



        try:
            name = driver.find_elements(By.XPATH,'//div[@class="Table_dataCell__2QCve"]')
            for i in name[::4]:
                name_of_person.append(i.text)

        except NoSuchElementException:
            name_of_person.append("NA")

        try:
            rank = driver.find_elements(By.XPATH,'//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]')
            for i in rank:
                rank_of_person.append(i.text)
        except NoSuchElementException:
            rank_of_person.append("-")

        try:
            age = driver.find_elements(By.XPATH,'//div[@class="Table_dataCell__2QCve"]')
            for i in age[1::4]:
                age_person.append(i.text)
        except NoSuchElementException:
            age_person.append("-")


        try:
            net = driver.find_elements(By.XPATH,'//div[@class="Table_netWorth___L4R5 Table_dataCell__2QCve"]')
            for i in net:
                net_worth_person.append(i.text)

        except NoSuchElementException:
            net_worth_person.append("-")


        try:
            citizen = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"]')
            for i in citizen[4::7]:
                citizenship_person.append(i.text)

        except NoSuchElementException:
            citizenship_person.append("-")

        try:
            source = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"]')
            for i in source[5::7]:
                source_person.append(i.text)

        except NoSuchElementException:
            source_person.append("-")

        try:
            industry = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"]')
            for i in source[6::7]:
                industry_person.append(i.text)

        except NoSuchElementException:
            industry_person.append("-")


        try:
            next_button = driver.find_element(By.XPATH,'//span[@class="Pagination_bubbleArrow__WFrX4 Pagination_paginationBtnNext__IOwqm"]')
            next_button.click()
        except ElementClickInterceptedException:
            print("End of The Page")
except NoSuchElementException:
    print("No Such Element Found")
    
except ElementClickInterceptedException:
    print("Page is end")
    
    
driver.quit()
```

```python
lets_go = driver.find_element(By.XPATH,'//button[@id="search-button"]')
lets_go.click()
```

```python
hostel_name = []
distance_from_city = []
ratings = []
total_reviews = []
overall_reviews=  []
private_price = []
dorms_price = []
facilities = []
property_desc  = []
opening_url = []
private_price_2 = []


try:
    name = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]/a')
    for i in name:
        hostel_name.append(i.text)
    for j in name:
        opening_url.append(j.get_attribute('href'))
except NoSuchElementException:
    print("End")
    
try:
    distance = driver.find_elements(By.XPATH,'//span[@class="description"]/span')
    for i in distance:
        distance_from_city.append(i.text.split("-")[1].strip())
except NoSuchElementException:
    distance_from_city.append("-")


try:
    rate = driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]/div[1]')
    for i in rate:
        ratings.append(i.text.split("\n")[0])
        
except NoSuchElementException:
    ratings.append("-")
    
    
try:
    total_re = driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for i in total_re:
        total_reviews.append(i.text.split(" ")[0])
except:
    total_reviews.append("-")

    

try:
    overall = driver.find_elements(By.XPATH,'//div[@class="keyword"]')
    for i in overall:
        overall_reviews.append(i.text)
except NoSuchElementException:
    overall_reviews.append("-")

    
try:
    private_p = driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    for i in private_p[::2]:
        
        private_price.append(i.text.split("Rs"))
    for j in private_price:
        if len(j)==3:
            private_price_2.append(j[2])
        elif len(j)==2:
            private_price_2.append(j[1])
        elif len(j)==1:
            private_price_2.append(j[0])
        
        
except NoSuchElementException:
    private_price.append("-")
dorm_p = []

try:
    dorm = driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    for i in dorm[1::2]:
        dorm_p.append(i.text.split("Rs"))
    for j in dorm_p:
        if len(j)==3:
            dorms_price.append(j[2])
        elif len(j)==2:
            dorms_price.append(j[1])
        elif len(j)==1:
            dorms_price.append(j[0])
        
except NoSuchElementException:
    dorm_p.append("-")
        
try:
    faci = driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')
    for i in faci:
        facilities.append(i.text.replace("\n"," "))
except NoSuchElementException:
    facilities.append("-")

time.sleep(2)

try:
    for i in opening_url:
        driver.get(i)
        show_more = driver.find_element(By.XPATH,'//a[@class="toggle-content"]')
        time.sleep(2)
        show_more.submit
        desc = driver.find_element(By.XPATH,'//div[@class="collapsible"]/div[2]')
        property_desc.append(desc.text)
    
except NoSuchElementException:
    property_desc.append("-")
```

```python
df = pd.DataFrame({"Hostel Name": hostel_name,
                  "Distance":distance_from_city,
                  "Rating in Star(Out 0f 10)":ratings,
                  "Total Reviews": total_reviews,
                  "Overall Reviews": overall_reviews,
                  "Private Room Cost": private_price_2,
                  "Dorms Cost": dorms_price,
                  "Facilities": facilities,
                  "Property Description": property_desc})
```
